Plot Google Smart TLS usage statistics + 7 day moving average

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn # make figures less ugly
```

```python
ts = pd.read_csv('https://www.google.com/transparencyreport/saferemail/google-starttls-percentages.csv', index_col=0)
```

```python
ts.index = pd.to_datetime(ts.index)
```

```python
ts.plot()
plt.title('Google Smart TLS usage statistics')
plt.show()
```

```python
ts.head()
```

```python
pd.rolling_mean(ts[['Inbound', 'Outbound']], window=7).plot()
plt.title('Google Smart TLS usage statistics (7-day avg)')
plt.show()
```
